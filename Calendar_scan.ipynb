# Import data and models

```python
import utils2
X_train, X_test, y_train, y_test, loaded_data = utils2.load_shuffle_split_scale()
X_train, X_test, y_train, y_test = utils2.upsampling(X_train, X_test, y_train, y_test)
reg_pred,svc_pred,RanFor_pred = utils2.generate_models(X_train, X_test, y_train, y_test,show_graphs=False)
positives = utils2.create_trade_list(reg_pred,svc_pred,RanFor_pred,loaded_data,X_test,y_test,show_graphs=False) 
```

```python
th,inv = utils2.user_threshold()
trades_taken,th,inv,summary = utils2.threshold_trades(positives,th,inv)
trades_taken = utils2.resulting_investment(trades_taken,th,inv,summary)
```
